### Lets see How a normal neuron works in Feed Forward Network



### `Recurrent Neuron` is little different
* It sends output back to itself!



### We can unfold in time, `RNN` takes input from previous input and current input as well

* Cells that are a function of inputs from previous time steps are also know as memory cells.
* `RNN` are aslo flexible in thier inputs and outputs, for both sequences and single vector values.

* We can also create entire layers of Recurrent Neurons
* RNN layer with 3 neurons
* Unroll layer thourghout time
* Part of this network which hold the information is know as memory cell.



# `LSTMs, GRU, Text Generation`

* An issue RNN face is that after awhile the network will begin to "forget" the first inputs, as information is lost at each step goint through the RNN.
* We need some sort of "Long term memory" for our network.

* The LSTM (Long Short Term Memory) cell was created to help adress these RNN issues.

#### The very first step is forget gate layer and here we decide what informtion to forget for the cell state
* when we pass in ${H}_{t-1}$ and ${X}_{t}$ after performing linear transformation with some weights and bias term into a sigmoid function and if we remeber sigmoid function outputs 0 and 1 where 1 reprsent to keep the information and 0 to forget

* for example if we are working with Language model where we want to predict next word based on previous one a cell state might include a gender of the present subject object, so that we end picking the correct pronoun. When we are picking up the new subject we want to forget the gender of old subject and that will be use of forget gate layer

#### The second step is decide what informtion we wan to remember in cell state.

* First part is sigmoid layer and second part is hyperbolic tangent layer
* Sigmoid layer is know as input gate layer 
* Then hyperbolic tangent (tanh layer)
* That end up creating vector which is know as `new candidate values`

#### In the next step we gona combine these two to update to the cell state

* In example of Language model we essentially wants to add the gender of the new subject to the cell state and replace the old that we already decided we are forgetting

#### Then we updated the old cell state ${C}_{t-1}$ and update it will ${C}_{t}$ so that we can pass it on to  ${C}_{t+1}$

### For this to happen we multiply old state with forget gate then we add it with input gate layer multipled with new candidate values

#### If we think from prospect of language model this is  where we actually drop the information about that old subjects gender and add in new information based of what we decided in the previous steps 

### Out final decision will be what do we output for ${h}_{t}$, so this output will based on your cell state but it's just filtered version.

### There different variances to the LSTMs, one of them is `"peepholes"`



### This is quite popular as this add peepholes to all the gates, basically it allows all the gates to see the previous cell state

#  Another variation of LSTMs are `Gated Recurrent Unit(GRU)`



### It simplifies things a bit by combining forget and input gate into a single gate or know as `Update Gate`. It also merges the cell state and hidden state which results in simplete model

```python
def create_model(vocabulary_size, seq_len):
    model = Sequential()
    model.add(Embedding(vocabulary_size, seq_len, input_length = seq_len))## We are defining input dmension for embedding
                                                                          ## also output lenght as seq_len and input length
                                                                          ## which is same as output length
    # Embedding is actually turns positive integers(index) into dense vectors of fixed size
    # This has to be the first layer
    model.add(LSTM(50, return_sequences = True))
    model.add(LSTM(50))
    model.add(Dense(50, activation = 'relu'))
    
    model.add(Dense(vocabulary_size, activation = 'softmax'))
    model.compile(loss = "categorical_crossentropy", optimizer= 'adam', metrics = ['accuracy'])
    
    model.summary()
    
    return model
```

# ` Now we will se how to generate text`

```python
from keras.utils import pad_sequences
```

```python
def generate_text (model, tokennizer, seq_len, seed_text, num_gen_words):
    
    output_text =[]
    # output text is the final output
    #start with 25 word
    input_text = seed_text
    # seed_text is intial seeding text
    
    for i in range(num_gen_words):
        ## lets say num_gen_words = 10, we want to generate 10 words
        
        encoded_text = tokennizer.texts_to_sequences([input_text])[0]
        ## first take input text string and then encode it to sequence
        
        pad_encoded = pad_sequences([encoded_text], maxlen = seq_len, truncating="pre")
        # If seed text is too short or too long we may need to pad it or need to cut it off or may need to add to it
        
        predicted_word =np.argmax( model.predict(pad_encoded, verbose =0)[0])
        # predict the class probability for each word
        
        pre_word = tokennizer.index_word[predicted_word]
        # Actual predicted word
        
        input_text += ' '+pre_word
        # then we take the input text add space and add predicted word. It create 26 word
        
        output_text.append(pre_word)
    
    return ' '.join(output_text)
```

#### We can create a function to help tokenizing

```python
def  vectorize_stories(data, word_index = tokenizer.word_index, max_story_len = max_story_len, max_question_len = max_question_len):
    # stories
    X = []
    
    # questions
    Xq = []
    
    # Y Correct Answer(yes/no)
    
    Y = []
    
    for story, query, answer in data:
        
        # for each story
        # creating a list like [23, 14,... and so on ]
        x = [word_index[word.lower()] for word in story]
        # for each question
        xq = [word_index[word.lower()] for word in query]
        
        y = np.zeros(len(word_index) + 1)
        
        y[word_index[answer]] = 1
        
        X.append(x)
        Xq.append(xq)
        Y.append(y)
        
    ### pad these squences and create final tuple which is unpacked
    
    return (pad_sequences(X, maxlen= max_story_len), pad_sequences(Xq, maxlen= max_question_len),np.array(Y))
    
    
    
    
    
```

### We have to first create a place holder using Input, Input instanciate a keras tensor

```python
### THis is placeholder
## it will take shape = (max_story_len, batch_size)
## Batch_Size we haven't defined yet so we leave it blank
input_sequence = Input((max_story_len, ))
question = Input((max_question_len,))
```

### We have created placeholders for story and question which will take inputs in batches

#### We will now create the Input Encoders and first thing we will do is to define Vocab size

```python
vocab_size = len(vocab) + 1
```

#### First we will create Input Encoder M then Input Encoder C
##### This input gets embeded with sequences of vectors

```python
# input ecoder M
input_encoder_m = Sequential()
input_encoder_m.add(Embedding(input_dim = vocab_size, output_dim = 64))
input_encoder_m.add(Dropout(0.3))
## Dropout layer turns off percentage randomly of neurons as we are training , With the above we are setting 30% as dropout
## it essentially helps overfitting, higher value takes longer time to train
### This encoder going to outputs something in the form of (samples, story_maxlen, embedding_dim) 
```

```python
# input ecoder c
input_encoder_c = Sequential()
input_encoder_c.add(Embedding(input_dim = vocab_size, output_dim = max_question_len))
input_encoder_c.add(Dropout(0.3))
## Dropout layer turns off percentage randomly of neurons as we are training , With the above we are setting 30% as dropout
## it essentially helps overfitting, higher value takes longer time to train
### This encoder going to outputs something in the form of (samples, story_maxlen, max_question_len)) 
```

```python
### Now we will have the question encoder
question_encoder = Sequential()
question_encoder.add(Embedding(input_dim = vocab_size, output_dim = 64, input_length = max_question_len))
question_encoder.add(Dropout(0.3))
# output
#(Samples, query_maxlen, embedding_dim)
```

#### Now we have Input encoder M , input encoder C and Question Encoder . Now We will encode the Sequences.

#### As we already have place holder created for inputs when we created` input_sequence` and `question` with placeholder , So we will pass in these inputs in out input encoders.

```python
input_encoded_m = input_encoder_m(input_sequence)
input_encoded_c = input_encoder_c(input_sequence)
question_ecoded = question_encoder(question)
```

#### We now need to use a Dot product to compute the match between the first inout vetor sequence and the querry

```python
match = dot([input_encoded_m, question_ecoded], axes = (2,2))
match = Activation('softmax')(match)
```

#### Now we want to add this Match vector with second input_sequence

```python
response = add([match, input_encoded_c])
response = Permute((2,1))(response)
# Permutes the dimensions of the input according to a given pattern.
```

#### Once we have repsonse we can concatenate the match matrix with the question vector sequence

```python
answer = concatenate([response, question_ecoded])
```

```python
answer
```

#### Now we have our `answer` `tensor`, we will reduce it with recurrent neural network and we will specifically choose the LSTM layer to do this

```python
answer = LSTM(32)(answer)
```

#### And we will perform one more series of regularization with Dropout

```python
answer = Dropout(0.5)(answer)
answer = Dense(vocab_size)(answer)### this will out put (samples, vocab_size) with marking of and yes and no esle 
# everything will be buch of zeros
```

#### Now we need to output a probability distribution over the vocabulary, becaue we essentially see a buch of zeros, except some probability on yes and some probability on no and we pass it through `softmax` in order tto get in 0 and 1.

```python
answer = Activation('softmax')(answer)
```

### Now finally we can build out model

```python
model = Model([input_sequence,question], answer)
```

```python
model.compile(optimizer='rmsprop', loss= 'categorical_crossentropy', metrics= ['accuracy'])
```
